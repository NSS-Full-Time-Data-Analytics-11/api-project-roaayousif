```python
import requests
import json
import pandas as pd
```

1.Make an API request that returns the months where "fire" was searched in 2016. Which month had the most searches? 

```python
endpoint = 'https://data.nashville.gov/resource/fuaa-r5cm.json?$where=year=2016 AND query_text= "fire"'

response = requests.get(endpoint)
```

```python
response
```

```python
response.text
```

```python
fire_data = pd.DataFrame(response.json())
fire_data
```

```python
fire_data.sort_values('query_count', ascending=False).head(1)
```

2.Make an API request that returns all the times a query was run more than 100 times in a month. How many times did this occur? 

```python
endpoint = 'https://data.nashville.gov/resource/fuaa-r5cm.json?$where=query_count > 100'

response2 = requests.get(endpoint)
```

```python
response2.text
```

```python
query_runs = pd.DataFrame(response2.json())
query_runs
```

```python
len(query_runs)
```
